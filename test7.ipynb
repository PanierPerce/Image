# 实验七

使用python实现大津法和迭代阈值法，比较两者性能差别。

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('greygraph.png', 0)
T = img.mean()

while True:
    t0 = img[img < T].mean()
    t1 = img[img >= T].mean()
    t  = (t0 + t1) / 2
    if abs(T - t) < 1:
        break
    T = t
T = int(T)
```

```python
img = cv.imread('greygraph.png',0)
th, img_OTSU = cv.threshold(img, 0, 255, cv.THRESH_OTSU)
th, img_iter = cv.threshold(img, T, 255, 0)

plt.subplot(131), plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)),
plt.title('Origin')
plt.axis('off')
plt.subplot(132), plt.imshow(cv.cvtColor(img_OTSU, cv.COLOR_BGR2RGB)),
plt.title('OTSU')
plt.axis('off')
plt.subplot(133), plt.imshow(cv.cvtColor(img_iter, cv.COLOR_BGR2RGB)),
plt.title('Iteration')
plt.axis('off')
```
