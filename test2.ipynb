使用python编程实现图像的 ***平移***、***镜像***、***旋转*** ，以及三种几何变换的复合。

```python
from PIL import Image

img = Image.open("greygraph.png")
img
```

```python
img.size
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

cvimg = cv.imread("greygraph.png")
cvimg = cv.cvtColor(cvimg, cv.COLOR_BGR2RGB)
rows, cols, chnl = cvimg.shape
M = np.float32([[1,0,100],[0,1,100]]) # 向下、右平移
cvimg1 = cv.warpAffine(cvimg,M,(cols,rows))
plt.imshow(cvimg1)
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
out1 = img.transpose(Image.FLIP_LEFT_RIGHT)
out1 # 镜像
```

```python
out2 = img.resize((600,600))
out2 # 变形
```

```python
# out3 = img.rotate(45)
out3 = img.rotate(288)
out3 # 旋转
```

```python
from PIL import ImageChops


def ImgTranslate(img, xoff, yoff):
    width, height = img.size
    c = ImageChops.offset(img, xoff, yoff)
    c.paste((0, 0, 0), (0, 0, xoff, height))
    c.paste((0, 0, 0), (0, 0, width, yoff))
    return c

out5 = ImgTranslate(out1.rotate(140), 35, 62)
out5 # 复合操作
```
