# 实验一

使用python对随机一张灰度照片进行 ***线性***、***分段线性***、***非线性*** 的点运算，对比原图和进行灰度变化后的效果。

```python
from PIL import Image

img = Image.open("greygraph.png")
img # 原图
```

```python
out = img.point(lambda i: i*1.2+0.3)
out  # 整体变亮
```

```python
out2 = img.point(lambda i: i*0.6-0.8)
out2  # 整体变暗
```

```python
out3 = img.point(lambda i: i * (-1))
out3 # 亮区域变暗，暗区域变亮（？）
```

```python
out3 = img.point(lambda i: 255 - i)
out3 # 亮区域变暗，暗区域变亮
```

```python
out4 = img.point(lambda i: i*0.5 if i<120 else i*1.5)
out4 # 分段线性，一部分扩展一部分压缩
```

```python
import math
out5 = img.point(lambda i: math.log(1+i)*50)
out5 # 对数变换：低灰度区扩展，高灰度区压缩。图像加亮、减暗
```

```python
out6 = img.point(lambda i: math.pow(i,1.5))
out6 # 幂次变换：低灰度区扩展，高灰度区压缩。图像加亮、减暗
```
